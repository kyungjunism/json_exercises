Using data in file 'data/world_bank_projects.json' and the techniques demonstrated above,

```python
import pandas as pd
import json
from pandas.io.json import json_normalize
import numpy as np
```

```python
# Read and save dataset as a dataframe:
df = pd.read_json('data/world_bank_projects.json')

# Inspect the state of the dataframe:
df.info()
```

## 2. Find the top 10 major project themes (using column 'mjtheme_namecode').

```python
#Must load json file as string since df cannot handle nested lists within dataframes
df = json.load((open('data/world_bank_projects.json')))
df = json_normalize(df, 'mjtheme_namecode')
df.head()
```

```python
df =pd.DataFrame(df)
df.name.value_counts()
```

Must fill in the missing values for the complete answer

## 3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

We need to sort values by code so that we can properly align all that is missing with backfill

```python
df = df.sort_values(['code', 'name'])

df.name[df['name'] == ''] = np.nan
df.head(100)
```

```python
df.name.value_counts().head(10)
```

The top ten project themes are above!
